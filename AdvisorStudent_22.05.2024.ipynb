```python
import pyAgrum.lib.notebook as gnb
import pyAgrum as gum
import cProfile
import random
import sys
import pyAgrum.lib.ipython as gnbipy
sys.path.append("C:\\Users\\zlhyl\\OneDrive\\Belgeler\\PhD")


    
from SDM_Functions import (
    update_decision_node_options, 
    add_nodes_from_student_to_advisor, add_nodes_from_advisor_to_student, add_equal_probabilities_for_new_option,
    add_decision_related_arcs_from_advisor_to_student,
    add_decision_related_arcs_from_student_to_advisor2, add_utility_related_arcs_from_advisor_to_student,
    add_utility_related_arcs_from_student_to_advisor, add_chance_related_arcs_from_advisor_to_student,
    add_chance_related_arcs_from_student_to_advisor, list_cpts_of_model, list_unmatched_node_cpts,
    add_noise_to_cpt, fill_student_cpts_with_noisy_data, fill_advisor_cpts_with_noisy_data,
    print_cpts_for_comparison, has_empty_cpts, print_all_cpts, print_all_utilitiesXX,
    get_utilities_as_dict_with_names,add_utility_related_arcs_from_advisor_to_studentORJ,
    pool,selected_cpt_difficulty,
    selected_cpt_friends,selected_cpt_interest,add_cpt_from_pool,calculate_and_set_utilities_StudentS,calculate_and_set_utilities_AdvisorS,
    compare_decision_utilities,calculate_utilities_for_states,show_decisions,show_decision_utilities,compare_final_decisions,
    compare_final_decisions_panda,calculate_initial_results,print_all_utilitiesX,filter_chance_nodes,combine_influence_diagrams,
    add_nodes_with_percentage,show_decision_utilities3,
    get_utility_parents,find_matched_parents,find_unmatched_parents,get_parents_of_unmatched_parents,find_and_get_parents_of_unmatched_parents,
    check_unmatched_parents_in_expert_model,check_unmatched_parents_in_expert_model_final_situation,filter_chance_nodes,find_and_get_parents_of_matched_parents,
    get_parents_of_matched_parents,get_arcs_to_utility,delete_unmatched_arcs,remove_unwanted_arcs,transfer_utilities,filter_chance_nodes_ORJ,add_arc_to_influence_diagram,
    remove_arc_from_influence_diagram
)

from advisor_model_v04 import create_advisor_model
from student_model_v05 import create_student_model

advisor_model = create_advisor_model()
student_model = create_student_model()


initial_advisor_model = create_advisor_model()
initial_student_model = create_student_model()

gnb.sideBySide(advisor_model,student_model)

#utilities_model_student1, max_expected_utility_student,_ = show_decision_utilities3(student_model)                
#utilities_model_advisor1, max_expected_utility_advisor,_ = show_decision_utilities3(advisor_model)   
#result_df1 = compare_final_decisions(utilities_model_student1, utilities_model_advisor1, threshold=5)
#print(result_df1)

       
## functions for the student model
# 1- DECISION NODE TRANSFER
unmatched_parents_student=find_unmatched_parents(student_model,initial_advisor_model)
student_model,student_decision_node_name = update_decision_node_options(student_model,initial_advisor_model) 
#if random.random() < 0.5:
#    add_equal_probabilities_for_new_option(student_model)
 
#else:
add_cpt_from_pool(student_model, selected_cpt_difficulty, selected_cpt_friends)

        
# 2- CHANCE NODE TRANSFER
matched_parents_student=find_matched_parents(initial_student_model,student_model)
#advisor_model=delete_unmatched_arcs(advisor_model, student_model)
student_model, unmatched_nodes_student=add_nodes_from_advisor_to_student(student_model,initial_advisor_model)
##add arcs to the advisor model
student_model=add_utility_related_arcs_from_student_to_advisor(initial_advisor_model,student_model)
student_model=add_decision_related_arcs_from_advisor_to_student(initial_advisor_model,student_model)
student_model=add_chance_related_arcs_from_advisor_to_student(initial_advisor_model,student_model)
student_model = fill_student_cpts_with_noisy_data(student_model,advisor_model, unmatched_nodes_student)
#student_model=remove_unwanted_arcs(student_model, unmatched_parents_student)
#student_model=filter_chance_nodes(advisor_model, nodes_to_delete)

nodes_to_delete_2 = ["career_development","prerequisite"]
student_model=filter_chance_nodes(student_model, nodes_to_delete_2)
#student_model=remove_unwanted_arcs(student_model, unmatched_parents)
# Add an arc from node 'A' to node 'B'
add_arc_to_influence_diagram(advisor_model, 'course', 'Grade')
remove_arc_from_influence_diagram(student_model, 'difficulty', 'workload')

#gnb.sideBySide(advisor_model, student_model)

CPT1_grade = [    
    0.7, 0.2, 0.1
]

CPT2_grade = [    
    0.5, 0.1, 0.4
]

CPT3_grade = [    
    0.1, 0.3, 0.6
]

CPT4_grade = [    
    0.2, 0.5, 0.3
]

CPT5_grade = [    
    0.6, 0.3, 0.1
]

CPT1_friends = [    
    0.7, 0.3
]

CPT2_friends = [    
    0.6, 0.4
]

CPT3_friends = [    
    0.5, 0.5
]

CPT4_friends = [    
    0.2, 0.8
]

CPT5_friends = [    
    0.1, 0.9
]


pool = {
    "grade1": [CPT1_grade, CPT2_grade, CPT3_grade, CPT4_grade, CPT5_grade],
    "grade2": [CPT3_grade, CPT4_grade],
    "grade3": [CPT5_grade]}

selected_cpt_grade1 = random.choice(pool["grade1"])
selected_cpt_grade2 = random.choice(pool["grade2"])
selected_cpt_grade3 = random.choice(pool["grade3"])

def add_cpt_from_pool_grade(advisor_model, selected_cpt_grade1, selected_cpt_grade2, selected_cpt_grade3):
    advisor_model.cpt('Grade')[{'course': 'Course1'}] = selected_cpt_grade1
    advisor_model.cpt('Grade')[{'course': 'Course2'}] = selected_cpt_grade2
    advisor_model.cpt('Grade')[{'course': 'Course3'}] = selected_cpt_grade3

pool2 = {
    "friends1": [CPT1_friends, CPT2_friends, CPT3_friends, CPT4_friends, CPT5_friends],
    "friends2": [CPT1_friends, CPT2_friends, CPT3_friends, CPT4_friends, CPT5_friends],
    "friends3": [CPT1_friends, CPT2_friends, CPT3_friends, CPT4_friends, CPT5_friends]}

selected_cpt_friends1 = random.choice(pool2["friends1"])
selected_cpt_friends2 = random.choice(pool2["friends2"])
selected_cpt_friends3 = random.choice(pool2["friends3"])

def add_cpt_from_pool_friends(advisor_model, selected_cpt_friends1, selected_cpt_friends2, selected_cpt_friends3):
    advisor_model.cpt('friends')[{'course': 'Course1'}] = selected_cpt_friends1
    advisor_model.cpt('friends')[{'course': 'Course2'}] = selected_cpt_friends2
    advisor_model.cpt('friends')[{'course': 'Course3'}] = selected_cpt_friends3
    
    
## functions for the advisor model
unmatched_parents=find_unmatched_parents(advisor_model,student_model)
print(unmatched_parents_student)
advisor_model, advisor_decision_node_name = update_decision_node_options(advisor_model, student_model) 
matched_parents=find_matched_parents(student_model, advisor_model)
#advisor_model=delete_unmatched_arcs(advisor_model, student_model)
advisor_model, unmatched_nodes_advisor = add_nodes_from_student_to_advisor(advisor_model, student_model,unmatched_parents_student)  
##add arcs to the advisor model
advisor_model=add_utility_related_arcs_from_student_to_advisor(student_model,advisor_model)
advisor_model=add_decision_related_arcs_from_student_to_advisor2(student_model,advisor_model)
advisor_model=add_chance_related_arcs_from_student_to_advisor(student_model,advisor_model)
#advisor_model=fill_advisor_cpts_with_noisy_data(advisor_model, student_model, unmatched_parents_student)

#add_cpt_from_pool_grade(advisor_model,selected_cpt_grade1, selected_cpt_grade2, selected_cpt_grade3)
#advisor_model=remove_unwanted_arcs(advisor_model, initial_advisor_model,matched_parents)


## when we update the advisor model, copy the utilies from the student model
#***copy the utilities


# 3) UTILITY TRANSFER

# Find the utility node in the advisor model

# Set the name of the utility node
#fin_advisor_model.variable(utility_node).setName("Student Utility")

def add_utility_with_noise(advisor_model, mu=0, sigma=5):
    utilities = {}
    utility_node = next(node for node in advisor_model.nodes() if advisor_model.isUtilityNode(node))

    # Iterate over all possible combinations of states
    for grade_val in [">C", "D", "F"]:
        for fr_val in ["yes", "no"]:
            for wr_val in ["high", "medium", "low"]:
                # Original utility calculation
                v = 60 if fr_val == "yes" else 10
                if grade_val == ">C":
                    y = 50
                elif grade_val == "D":
                    y = 30
                else:
                    y = 0
                if wr_val == "high":
                    x = 20
                elif wr_val == "medium":
                    x = 30
                else:
                    x = 40

                # Apply noise to the utility calculation
                utility_value = 0.2 * v + 0.5 * y + 0.1 * x + generate_noise(mu, sigma)
                
                # Store the utility value for the combination of states
                utilities[(grade_val, fr_val, wr_val)] = utility_value

    # Set the utility values in the advisor model
    for (grade_val, fr_val, wr_val), utility_value in utilities.items():
        advisor_model.utility(utility_node)[{'Grade': grade_val, 'friends': fr_val, 'workload': wr_val}] = utility_value
import numpy as np

def generate_noise(mu, sigma):
    # Generate noise based on normal distribution
    noise = np.random.normal(mu, sigma)
    return noise

add_utility_with_noise(advisor_model, mu=0, sigma=5)



## delete unmatched parents of the advisor utility node
#advisor_model=filter_chance_nodes(advisor_model,unmatched_parents)

advisor_model=remove_unwanted_arcs(advisor_model, unmatched_parents)
add_arc_to_influence_diagram(advisor_model, 'course', 'Grade')
add_cpt_from_pool_grade(advisor_model,selected_cpt_grade1, selected_cpt_grade2, selected_cpt_grade3)
add_cpt_from_pool_friends(advisor_model,selected_cpt_friends1, selected_cpt_friends2, selected_cpt_friends3)
#gnb.sideBySide(advisor_model)

#print_all_utilitiesX(advisor_model)
gnb.sideBySide(advisor_model,student_model)
         
#print_all_cpts(advisor_model)
#print_all_cpts(student_model)


#utilities_model_student1, max_expected_utility_student,_ = show_decision_utilities3(student_model)                
#utilities_model_advisor1, max_expected_utility_advisor,_ = show_decision_utilities3(advisor_model)   
#result_df1 = compare_final_decisions(utilities_model_student1, utilities_model_advisor1, threshold=5)
#print(result_df1)


```
